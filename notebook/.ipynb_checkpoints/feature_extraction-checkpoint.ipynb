```python
import numpy as np
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg16 import preprocess_input
import os

# Load pretrained VGG16
base_model = VGG16(weights='imagenet', include_top=False)
model = Model(inputs=base_model.input, outputs=base_model.output)


def extract_features(generator, save_path):
    features = []
    labels = []

    print("\nStarting Feature Extraction...\n")

    for i in range(len(generator)):
        X_batch, y_batch = generator[i]

        # Convert back to 0–255 then preprocess
        X_batch = preprocess_input(X_batch * 255.0)

        # Extract features
        feat = model.predict(X_batch, verbose=0)

        # Flatten (7x7x512 → 25088)
        feat = feat.reshape(feat.shape[0], -1)

        features.append(feat)
        labels.append(y_batch)

        print(f"Processed batch {i+1}/{len(generator)}")

    X = np.vstack(features)
    y = np.vstack(labels)

    print("\nFeature extraction complete ✅")
    print("Shape:", X.shape)

    # Save features
    os.makedirs(save_path, exist_ok=True)

    np.save(os.path.join(save_path, "features.npy"), X)
    np.save(os.path.join(save_path, "labels.npy"), y)

    print(f"\nSaved features to: {save_path}")

    return X, y
```

```python
import sys
import os

# Add project root to path
sys.path.append(os.path.abspath("."))

# Import here (important)
from preprocessing import get_data_generators

BASE_PATH = r"D:\CODE\Alzheimer_Detection_And_Classification\data"

# Load generators
train_gen, val_gen, test_gen = get_data_generators(BASE_PATH)

# Extract features
X_train, y_train = extract_features(train_gen, "models/train_features")

print("\nDone ✅")
```
